```python
import networkx as nx
import matplotlib.pyplot as plt
from matching import Network
plt.rcParams['figure.figsize'] = [13.0, 8.0]
```

```python
def get_edgetrap(b=3):
    N = Network()

    N.add_link("s", "u0")
    N.add_link("s", "v0")

    counter = 1
        
    for i in range(b):
        N.add_matching_edge(f"v{i}", f"u{i}", b=counter)
        counter += 1
        N.add_matching_edge(f"v{i}", f"u{i+1}", b=counter)
        counter += 1
        N.add_link(f"u{i}", f"u{i+1}")
        N.add_link(f"v{i}", f"v{i+1}")
        N.add_matching_edge(f"v{i}", f"w{i}", b= counter)
        counter += 1
        N.add_link(f"u{i}", f"w{i}")
        N.add_matching_edge(f"w{i}", f"v{i+1}", b= counter)
        counter += 1

    N.add_matching_edge(f"v{b}", f"u{b}", b=counter)

    pos  ={"s":(-0.5,0.5)}
    for i in range(b+1):
        pos[f"v{i}"] = (i,1)
        pos[f"u{i}"] = (i,0)
        pos[f"w{i}"] = (i+1,2)
    return N, pos

N, pos = get_edgetrap(3)
```

```python
N.draw(pos)
```

```python
steps = 0 
while True:
    if not N.best_step():
        break
    steps += 1
N.draw(pos)
```

```python
iterations = []
for b in range(10):
    N, _  = get_edgetrap(b)
    iterations.append(N.iterate())

plt.plot(iterations)
```
